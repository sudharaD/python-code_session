```python
from sklearn.preprocessing import StandardScaler
import numpy as np

# Sample data (2-dimensional array)
data = np.array([[8678, 6],
                 [243, 876],
                 [43, 1456],
                 [26, 67],
                 [85, 123]])

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the data and transform the data
scaled_data = scaler.fit_transform(data)

print("Original Data:\n", data)
print("\nScaled Data:\n", scaled_data)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a numpy array with 5 numbers, where 3 of them are equal
my_array = np.array([2, 2, 3, 3, 3])

print(my_array) 
plt.scatter(scaled_data[:,0], scaled_data[:,1], c=my_array)
```
